## Task 3: Feature Engineering

Adding in the categorical features and encoding them correctly.

```python
X_train = X_f
y_train = y_f
```

```python
X_train.isnull().sum()
```

Dropping size since it is too missing

```python
categorical_f = ['region','manufacturer','model','condition','cylinders','fuel','title_status',
                'transmission','drive','type','paint_color','state']
X_train = X_train[continuous_f + categorical_f]
```

Going to treat missing data as another category.  Only one hot encoding low unique count features. Features 'region','manufacturer','model','state' will be target encoded.

```python
te_cols = ['region','manufacturer','model','state']
ohe = list(set(categorical_f) - set(te_cols))
```

```python
#target encode
te = TargetEncoder(cols=te_cols).fit(X_train, y_train)
te_cols_df = te.transform(X_train)

#missing cat data
imp1 = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='NA').fit(X_train[ohe])
X_train_ohe = imp1.transform(X_train[ohe])

#missing cont data
imp2 = SimpleImputer(strategy='median').fit(X_train[continuous_f])
X_train_cont = imp2.transform(X_train[continuous_f])

#convert back to dfs
ohe_cols_df = pd.DataFrame(X_train_ohe,columns=ohe)
cont_cols_df = pd.DataFrame(X_train_cont,columns=continuous_f)
te_cols_df = te_cols_df[te_cols].reset_index(drop=True)
```

```python
X_train_new = cont_cols_df.join(te_cols_df).join(ohe_cols_df)
#X_train_new.head()
```

```python
preprocess = make_column_transformer(
    (RobustScaler(), continuous_f),
    (OneHotEncoder(), ohe),
    remainder='passthrough'
    )

lin_pipe2 = make_pipeline(preprocess, LinearRegression())

lin_scores2 = cross_val_score(lin_pipe2, X_train_new, y_train, cv=10, scoring='neg_root_mean_squared_error')
lin_scores2.mean()
```

Adding polynomial features

```python
lin_pipe3 = make_pipeline(preprocess, PolynomialFeatures(degree=2), LinearRegression())

lin_scores3 = cross_val_score(lin_pipe3, X_train_new, y_train, cv=10, scoring='neg_root_mean_squared_error')
lin_scores3.mean()
```

It looks like adding the encoded features improved the model.  Also, adding polynomial features improved the model, but very slightly.  It is likely only some polynomial features are significant and can potentially drop the insignificant ones.

## Task 4: Any Model

```python
pipe = make_pipeline(preprocess, LinearSVR(max_iter=5000))
```

```python
param_grid_lsvm = {'linearsvr__C': np.logspace(-3, 2, 6),
              'linearsvr__loss': ['epsilon_insensitive','squared_epsilon_insensitive']}
grid = GridSearchCV(pipe, param_grid=param_grid_lsvm, cv=10)
best_lsvm = grid.fit(X_train_new, y_train)

print(best_lsvm.best_params_)
print("Best Linear SVM val score:", best_lsvm.best_score_)
```
